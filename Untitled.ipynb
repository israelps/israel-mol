```python
def clean_newtonX_file(uri):
    f = open(uri)
    data = f.readlines()
    f.close()
    
    import re
    data = [line.strip() for line in data]
    data = [re.sub(r' +', r' ',line)+'\n' for line in data]
    new_uri = uri.split('.')
    new_uri[1] = '-filtered.dat'
    new_uri = ''.join(new_uri)
    f = open(new_uri, 'w')
    f.writelines(data)
    f.close()
    print('file created at: %s'%new_uri)
    
    
```

```python
f = open('b-caroten/spectra/cross-section.dat')
cross_section = f.readlines()
f.close()
f = open('b-caroten/spectra/emission-rate.dat')
emission_rate = f.readlines()
f.close()
```

```python
import re

cross_section = [line.strip() for line in cross_section]
emission_rate = [line.strip() for line in emission_rate]

cross_section = [re.sub(r' +', r' ',line)+'\n' for line in cross_section]
emission_rate = [re.sub(r' +', r' ',line)+'\n' for line in emission_rate]
```

```python
f = open('b-caroten/spectra/cross-section-filtered.dat', 'w')
f.writelines(cross_section)
f.close()
f = open('b-caroten/spectra/emission-rate-filtered.dat', 'w')
f.writelines(emission_rate)
f.close()
```
