```python
def clean_newtonX_file(uri):
    f = open(uri)
    data = f.readlines()
    f.close()
    
    import re
    data = [line.strip() for line in data]
    data = [re.sub(r' +', r' ',line)+'\n' for line in data]
    new_uri = uri.split('.')
    new_uri[1] = '-filtered.dat'
    new_uri = ''.join(new_uri)
    f = open(new_uri, 'w')
    f.writelines(data)
    f.close()
    print('file created at: %s'%new_uri)
    
    
```

```python
f = open('b-caroten/spectra/cross-section.dat')
cross_section = f.readlines()
f.close()
f = open('b-caroten/spectra/emission-rate.dat')
emission_rate = f.readlines()
f.close()
```

```python
import re

cross_section = [line.strip() for line in cross_section]
emission_rate = [line.strip() for line in emission_rate]

cross_section = [re.sub(r' +', r' ',line)+'\n' for line in cross_section]
emission_rate = [re.sub(r' +', r' ',line)+'\n' for line in emission_rate]
```

```python
f = open('b-caroten/spectra/cross-section-filtered.dat', 'w')
f.writelines(cross_section)
f.close()
f = open('b-caroten/spectra/emission-rate-filtered.dat', 'w')
f.writelines(emission_rate)
f.close()
```

```python
clean_newtonX_file('lutein/spectra/abs/cross-section.dat')
clean_newtonX_file('neoxanthin/spectra/abs/cross-section.dat')
```

```python
b_caroten_abs = pd.read_csv('b-caroten/spectra/cross-section-filtered.dat', header=0, names=['Energy', 'Lambda', 'Cross Section', 'Err_Abs'], index_col='Energy', sep=' ')
b_caroten_emi = pd.read_csv('b-caroten/spectra/emission-rate-filtered.dat', header=0, names=['Energy', 'Lambda', 'Emission Rate', 'Err_Emi'], index_col='Energy', sep=' ')
b_caroten = pd.concat([b_caroten_abs, b_caroten_emi], sort=False)

cla_abs = pd.read_csv('chlorophyll-a/spectra/abs/cross_section.lx', header=0, delimiter=' ', names=['Energy', 'Cross Section'], index_col='Energy')
cla_emi = pd.read_csv('chlorophyll-a/spectra/emi/differential_rate.lx', header=0, delimiter=' ',  names=['Energy', 'Emission Rate'], index_col='Energy')
cla = pd.concat([cla_abs, cla_emi], sort=False)
cla['lambda(nm)'] = 1024/cla.index

ph_bilin_abs = pd.read_csv('phycocyanobilin/spectra/abs/cross_section.lx', header=0, delimiter=' ', names=['Energy', 'Cross Section'], index_col='Energy')
ph_bilin_emi = pd.read_csv('phycocyanobilin/spectra/emi/differential_rate.lx', header=0, delimiter=' ',  names=['Energy', 'Emission Rate'], index_col='Energy')
phycocyanobilin = pd.concat([ph_bilin_abs, ph_bilin_emi], sort=False)
phycocyanobilin['lambda(nm)'] = 1024/phycocyanobilin.index

lutein_abs = pd.read_csv('lutein/spectra/abs/cross-section-filtered.dat', header=0, names=['Energy', 'Lambda', 'Cross Section', 'Err_Abs'], index_col='Energy', sep=' ')
lutein_emi = pd.read_csv('lutein/spectra/emi/differential_rate.lx', header=0, delimiter=' ',  names=['Energy', 'Emission Rate'], index_col='Energy')
lutein = pd.concat([lutein_abs, lutein_emi], sort=False)
lutein['lambda(nm)'] = 1024/lutein.index

neoxanthin_abs = pd.read_csv('neoxanthin/spectra/abs/cross-section-filtered.dat', header=0, names=['Energy', 'Lambda', 'Cross Section', 'Err_Abs'], index_col='Energy', sep=' ')
#neoxanthin_emi = pd.read_csv('neoxanthin/spectra/emi/differential_rate.lx', header=0, delimiter=' ',  names=['Energy', 'Emission Rate'], index_col='Energy')
#neoxanthin = pd.concat([neoxanthin_abs, neoxanthin_emi], sort=False)
#neoxanthin['lambda(nm)'] = 1024/neoxanthin.index

zeaxanthin_abs = pd.read_csv('zeaxanthin/spectra/abs/cross_section.lx', header=0, delimiter=' ', names=['Energy', 'Cross Section'], index_col='Energy')
zeaxanthin_emi = pd.read_csv('zeaxanthin/spectra/emi/differential_rate.lx', header=0, delimiter=' ',  names=['Energy', 'Emission Rate'], index_col='Energy')
zeaxanthin = pd.concat([zeaxanthin_abs, zeaxanthin_emi], sort=False)
zeaxanthin['lambda(nm)'] = 1024/zeaxanthin.index

dihydrobiliverdin_abs = pd.read_csv('dihydrobiliverdin/spectra/abs/cross_section.lx', header=0, delimiter=' ', names=['Energy', 'Cross Section'], index_col='Energy')
dihydrobiliverdin_emi = pd.read_csv('dihydrobiliverdin/spectra/emi/differential_rate.lx', header=0, delimiter=' ',  names=['Energy', 'Emission Rate'], index_col='Energy')
dihydrobiliverdin = pd.concat([dihydrobiliverdin_abs, dihydrobiliverdin_emi], sort=False)
dihydrobiliverdin['lambda(nm)'] = 1024/dihydrobiliverdin.index

dihydrobiliverdin = dihydrobiliverdin.set_index('lambda(nm)')
dihydrobiliverdin.head()
```

```python
plt.figure(figsize=(8,5))


ax1 = dihydrobiliverdin['Cross Section'].plot(color='blue', grid=True, label='Absorption Spectra')
ax2 = dihydrobiliverdin['Emission Rate'].plot(color='red', grid=True, secondary_y=True, label='Emission Rate')

h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.set_ylabel('Cross Section (A)')
ax2.set_ylabel('Emission Rate (eV-1)')

ax2.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)

plt.legend(h1+h2, l1+l2, loc=1)
plt.xlabel = 'Wave Length (nm)'
ax1.set_xlim([0, 2500])
ax1.set_ylim([0,1.8])
ax2.set_ylim([0,2*10**(-10)])

plt.savefig('dihydrobiliverdin/spectra/spectra.png', dpi=600)
plt.show()
```
