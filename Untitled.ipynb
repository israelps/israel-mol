```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
f = open('b-caroten/spectra/cross-section.dat')
cross_section = f.readlines()
f.close()
f = open('b-caroten/spectra/emission-rate.dat')
emission_rate = f.readlines()
f.close()
```

```python
import re

cross_section = [line.strip() for line in cross_section]
emission_rate = [line.strip() for line in emission_rate]

cross_section = [re.sub(r' +', r' ',line)+'\n' for line in cross_section]
emission_rate = [re.sub(r' +', r' ',line)+'\n' for line in emission_rate]
```

```python
f = open('b-caroten/spectra/cross-section-filtered.dat', 'w')
f.writelines(cross_section)
f.close()
f = open('b-caroten/spectra/emission-rate-filtered.dat', 'w')
f.writelines(emission_rate)
f.close()
```

```python
b_caroten_abs = pd.read_csv('b-caroten/spectra/cross-section-filtered.dat', header=0, names=['Energy', 'Lambda', 'Cross Section', 'Err'], index_col='Energy', sep=' ')
b_caroten_emi = pd.read_csv('b-caroten/spectra/emission-rate-filtered.dat', header=0, names=['Energy', 'Lambda', 'Emission Rate', 'Err'], index_col='Energy', sep=' ')
b_caroten_abs.head()
```

```python
hartree_to_ev = lambda x: 27.211*x
s0_homo = hartree_to_ev(-0.16381)
s0_lumo = hartree_to_ev(-0.07754)
s1_homo = hartree_to_ev(-0.15404)
s1_lumo = hartree_to_ev(-0.08724)
s0_homo, s0_lumo
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))


ax1 = b_caroten_abs['Cross Section'].plot(color='blue', grid=True, label='Absorption Spectra')
ax2 = b_caroten_emi['Emission Rate'].plot(color='red', grid=True, secondary_y=True, label='Emission Rate')

h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.set_ylabel('Cross Section (A)')
ax2.set_ylabel('Emission Rate (eV-1)')
ax1.set_xlim([1.32, 2.9])
ax1.set_ylim([0,10])
ax2.set_ylim([0,10*10**(-7)])
ax2.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)

plt.legend(h1+h2, l1+l2, loc=1)
plt.xlabel = 'Energy (eV)'
plt.savefig('b-caroten/spectra/spectra.png', dpi=600)
plt.show()
```
